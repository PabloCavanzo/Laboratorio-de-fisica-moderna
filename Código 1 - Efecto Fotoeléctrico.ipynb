```python
filea = '/content/drive/MyDrive/Colab/Lab de Moderna/Exp. 1 - Efecto fotoelectrico/Datos_EfectoFotoelectrico - Parte 1 (Amarillo) (1).csv'
```

```python
data_yellow = pd.read_csv(filea, usecols=['V(V)', 'I(nA)'])
data_yellow = data_yellow[1:]
data_yellow
```

```python
current_yellow = data_yellow["I(nA)"]
potential_yellow = data_yellow["V(V)"]
errores_amarillo = np.full(len(current_yellow), 0.001)

plt.errorbar(potential_yellow, current_yellow, yerr=errores_amarillo, fmt='o', color='y', capsize=3)
plt.scatter(potential_yellow, current_yellow, color='y')
plt.xlabel("Voltaje (V)")
plt.ylabel("Corriente (nA)")
plt.title("Voltaje vs Corriente LED Amarillo")
plt.legend(["Datos LED Amarillo"])
plt.grid(alpha=0.6)
```
